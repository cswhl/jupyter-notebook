```python
num = 2
# 将整数转换为bytes
print(str(num).encode('utf8'))  # 将int转换为bytes
print(bytes([2])) # 将int转为bytes

import struct
print(struct.pack('B', 2))
print(struct.pack('>H', 2))  # 大端模式
print(struct.pack('<H', 2))
print((2).to_bytes(2, byteorder="little"))
print((2).to_bytes(2, byteorder="big"))
```
